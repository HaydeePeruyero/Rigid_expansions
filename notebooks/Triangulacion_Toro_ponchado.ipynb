```python
def triang_toro(ponchaduras, resolucion):
    """
    Triangulación del toro con n ponchaduras
    Diagrama de un toro con n ponchadura y resolución r, la letra x simboliza la ponchadura
    Esta pieza es la pieza modular, en caso de tener dos ponchaduras, se agrega una pieza igual y la segunda ponchadura
    quedaría ahora con vértices 73,74,80,81. La cantidad total de vértices sería ahora es (7**2)*n

    0--7--14--21--28--35--42--0
    |  |   |   |   |   |   |  |
    1--8....................--1
    |                         |
    2--9....................--2
    |                          |
    3--10--17--24--31--38--45--3
    |  |    |   | x |   |  |   |
    4--11--18--25--32--39--40--4
    |                         |
    5--12...................--5
    |                          |
    6--13--20--27--34--41--48--6
    |                          |
    0--7--14--21--28--35--42--0
    """
    n = ponchaduras
    t = resolucion
    r = 2*t + 1
    # Indicar el número de ponchaduras
    # Añadir vertices de ponchaduras y lados
    vertices = []
    # añadir vertice de las esquinas
    vert_esq = [0]
    for i in range(len(vert_esq)):
        vertices.append(i)
    vert_ponch = []
    # vertices extremos verticales
    vert_lv = []
    ivv = 1
    for j in range(ivv,ivv+(r-1)):
        vert_lv.append(j)
        vertices.append(j)
    # añadir todos los vertices horizontales
    vert_lh = []
    ivh = r
    for j in range(ivh,ivh*ivh*(n)):
        vert_lh.append(j)
        vertices.append(j)
    # añadir aristas
    aristas = []
    for i in range(0, len(vert_lh), r):
    # Dividir los vértices de cada lado
        li = vert_lh[i:i+r]
    #Ex: [7,8,9,10,11,12,13,7]
        # aristas: [(7,8),(8,9),..,(12,13)]
    # Iterar sobre los elementos del bloque actual para crear las aristas
        for j in range(len(li) - 1):
            aristas.append((li[j], li[j + 1]))
    
    # Agregar la arista que conecta el último elemento con el primero del bloque [(13,7)]
        aristas.append((li[-1], li[0]))
    # añadimos las aristas de los extremos verticales [(0,1),(1,2),...(6,0)]
    aux1 = [0]
    for i in range(len(vert_lv)):
        aux1.append(vert_lv[i])
    for j in range(len(aux1)-1):
        aristas.append((aux1[j], aux1[j+1]))
    aristas.append((aux1[-1],aux1[0]))

    # Añadimos aristas horizontales (0,7),(7,14), (14,21),...,(42,0)
    aux2 = [0]
    for i in range(r*n):
        aux2.append(r*i)
    for j in range(len(aux2)-1):
        aristas.append((aux2[j], aux2[j+1]))
    aristas.append((aux2[-1],aux2[0]))
    #
    for k in range(1,r):
        aux3 = []
        for i in range(r*n):
            aux3.append(r*i+k)
        for j in range(len(aux3)-1):
            aristas.append((aux3[j], aux3[j+1]))
        aristas.append((aux3[-1],aux3[0]))

    aristas.remove((0,0))

    ponch_v = []
    for p in range(n):
        aux4 = []
        aux4 = [t*r + t + (r**2*p), (t+1)*r + t +(r**2*p), (t+1)*r + (t+1) + (r**2*p), t*r + (t+1) + (r**2*p)]
        ponch_v.extend(aux4)
        #ponch_ar = []

    a, b, c, d = 0, r, r+1, 1
    cubos_aux = [(a, b, c, d)]

    for j in range(r*n-2):
        aux8 = []
        aux8 = [(cubos_aux[0][0]+ r*(j+1), cubos_aux[0][1]+ r*(j+1), cubos_aux[0][2]+ r*(j+1), cubos_aux[0][3]+ r*(j+1))]
        cubos_aux.extend(aux8)

    aux9 = []
    aux9 = [(cubos_aux[0][0] + r*(r-1) + r**2*(n-1), cubos_aux[0][0], cubos_aux[0][3], cubos_aux[-1][2])]
    cubos_aux.extend(aux9)
    
    for i in range(1,r-1):
        aux5 = []
        aux5 = [(a+i, b+i, c+i, d+i)]
        #print(aux5)
        cubos_aux.extend(aux5)
        for j in range(r*n-2):
            aux6 = []
            aux6 = [(aux5[0][0]+ r*(j+1), aux5[0][1]+ r*(j+1), aux5[0][2]+ r*(j+1), aux5[0][3]+ r*(j+1))]
            #print(aux6)
            cubos_aux.extend(aux6)
        aux7 = []
        aux7 = [(aux5[0][0] + r*(r-1) + r**2*(n-1), aux5[0][0], aux5[0][3], cubos_aux[-1][2])]
        #print(aux7)
        cubos_aux.extend(aux7)
    cubos_aux.extend([(a+(r-1), b+(r-1), b, a)])

    for j in range(r*n-2):
        aux10 = []
        aux10 = [(a+(r-1)+ r*(j+1), b+(r-1)+ r*(j+1), cubos_aux[0][1]+ r*(j+1), cubos_aux[0][3]+ r*(j+1)-1)]
        cubos_aux.extend(aux10)

    aux11 = []
    aux11 = [(cubos_aux[-1][1], a+(r-1), a, cubos_aux[-1][2])]
    cubos_aux.extend(aux11)

    cubos = [
        tupla for tupla in cubos_aux
        if sum(1 for elem in tupla if elem in ponch_v) < 4
    ]
        
    return vertices, aristas, vert_lv, vert_lh, ponch_v, cubos
```

```python
def visualizacion(vertices, vertices_ponchaduras, ponchaduras, resolucion, subdivision=0):
    # Parámetros
    n = ponchaduras
    r = 2 * resolucion + 1

    # Dividir la lista en sublistas de tamaño 2n + 1
    sublistas = [vertices[i:i + r] for i in range(0, len(vertices), r)]

    # Diccionario para almacenar las coordenadas
    coordenadas = {}

    # Convertir cada sublista en el formato deseado
    for i, sublista in enumerate(sublistas):
        x_offset = i  # Incremento de x para cada sublista
        for j, elemento in enumerate(sublista):
            coordenadas[elemento] = (x_offset, j)

    # Añadir los vértices adicionales
    esquinas = {
        "0a": (0, r),
        "0c": (n * r, 0),
        "0b": (n * r, r)
    }
    coordenadas.update(esquinas)

    # Filtrar los primeros 2n vértices del primer paquete que tienen x = 0 (excluyendo el (0, 0))
    vertices_horizontales = [coordenadas[key] for key in range(1, r) if coordenadas[key][0] == 0]

    # Crear y agregar nuevos vértices con coordenadas horizontales
    for idx, coord in enumerate(vertices_horizontales, start=1):
        new_key = f"{idx}'"
        new_value = (n * r, coord[1])
        coordenadas[new_key] = new_value 

    # Filtrar los vértices donde y = 0 (excluyendo (0, 0))
    vertices_verticales = [key for key in coordenadas if coordenadas[key][1] == 0 and key != 0 and coordenadas[key] != (n * r, 0)]

    # Crear y agregar nuevos vértices con coordenadas verticales
    for key in vertices_verticales:
        new_key = f"{key}'"
        new_value = (coordenadas[key][0], r)
        coordenadas[new_key] = new_value

    coordenadas_unicas = {}
    valores_vistos = set()

    for key, value in coordenadas.items():
        if value not in valores_vistos:
            valores_vistos.add(value)
            coordenadas_unicas[key] = value

    #Almacenar las coordenadas de los cuadrados en una lista
    #cuadrados = []
    #for i in range
            
    # Graficar las coordenadas
    x = [coordenada[0] for coordenada in coordenadas.values()]
    y = [coordenada[1] for coordenada in coordenadas.values()]
    etiquetas = list(coordenadas.keys())

    plt.figure(figsize=(max(x)+3, max(y)+2))
    plt.scatter(x, y, color='blue')  # Dibujar los puntos

    # Colorear los cuadrados
    for i in range(0, len(vertices_ponchaduras), 4):
        if i + 3 < len(vertices_ponchaduras):  # Asegurarse de que hay cuatro vértices
            square_vertices = [coordenadas[vertices_ponchaduras[j]] for j in range(i, i + 4)]
            square_x = [coord[0] for coord in square_vertices]
            square_y = [coord[1] for coord in square_vertices]
            plt.fill(square_x, square_y, color='gray', alpha=0.5)  # Colorear el cuadrado en gris

    # Añadir etiquetas a cada punto
    for i, etiqueta in enumerate(etiquetas):
        plt.text(x[i] + 0.1, y[i] + 0.1, str(etiqueta), fontsize=12, ha='left', va='bottom')

    # Dibujar líneas horizontales y verticales entre los puntos
    # Agrupar puntos por fila
    filas = {}
    for i in range(len(x)):
        if y[i] not in filas:
            filas[y[i]] = []
        filas[y[i]].append(x[i])

    # Dibujar líneas horizontales
    for y_value, x_values in filas.items():
        plt.plot(x_values, [y_value] * len(x_values), color='gray', linestyle='--')

    # Agrupar puntos por columna
    columnas = {}
    for i in range(len(y)):
        if x[i] not in columnas:
            columnas[x[i]] = []
        columnas[x[i]].append(y[i])

    # Dibujar líneas verticales
    for x_value, y_values in columnas.items():
        plt.plot([x_value] * len(y_values), y_values, color='gray', linestyle='--')

    # Ajustes del gráfico
    plt.xlim(min(x)-2, max(x)+2)
    plt.xlim(min(y)-2, max(y)+2)
    plt.title(f'Toro con {ponchaduras} ponchadura(s) y resolución {r}')
    #plt.grid(True)
    plt.axis('equal')
    plt.xticks([])  # Quitar ticks en el eje x
    plt.yticks([])
    plt.show()
```
